```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn.utils
from sklearn.metrics import recall_score, accuracy_score, precision_score, f1_score, confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV
```

```python
df= pd.read_csv('Data_for_UCI_named.csv')
df
```

```python
df.describe
```

```python
df.info()
```

```python
X = df.drop(columns = 'stabf')
y = df['stabf']
scaler = StandardScaler()
X_norm = pd.DataFrame(scaler.fit_transform(X), columns = X.columns)
X_norm.head()
                      
```

```python
y.value_counts()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_norm,y, test_size = 0.2, random_state = 1)
```

###  QUESTION 14 

```python
random_f_clf = RandomForestClassifier(random_state=1)
random_f_clf.fit(X_train, y_train)
random_f_pred = random_f_clf.predict(X_test)
```

### accuracy :


```python
random_f_pred = random_f_clf.predict(X_test)

accuracy = accuracy_score(y_true=y_test, y_pred=random_f_pred)
print(f'Accuracy: {accuracy:.4f}')
```
